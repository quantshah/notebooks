```julia
using PyPlot
```

```julia
function ShowCircuit(N::Int32, gate_dict::Dict)
    """
    A simple implementation of quantum circuit visualization.
    
    Parameters
    ----------
    N : Int32
        The number of qubits in the circuit
    
    gate_dict : Dict
        A dictionary containing the position of each gate and the gate
        label as a key value pairs. (Tuple => AbstractString) 
    """
    figure(figsize=(10, 5));
    title_str = string("Displaying circuit with ",N," qubits");
    title(title_str);

    
    for position in keys(gates)
        gate_label = string("\$",gates[position],"\$");
        scatter(position[1],position[2], marker=gate_label, s=1000.,);
    end
    yticks(collect(1:N+1));
    x_len = maximum(maximum(keys(gates)))
    xticks([1:x_len]);
    grid(axis="x")
    grid(axis="y", linewidth=1, linestyle="-")
end
```

```julia
gates = Dict((1,1)=>"H", (2,2)=>"X", (1,3)=>"H", (6,1)=>"H", (4,4)=>"H")
ShowCircuit(3, gates)
```
