```python
from qutip import QubitCircuit
import networkx as nx
from qutip.qip.icm import Icm
from icm_graph import (construct_icm_graph, combine_graph, teleport,
                       three_loop_reduction, two_loop_reduction, draw_graph)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
circuit = QubitCircuit(2, reverse_states=False)

circuit.add_gate("CNOT", targets =[1], controls=[0])
circuit.add_gate("CNOT", targets =[0], controls=[1])
circuit.add_gate("CNOT", targets =[1], controls=[0])
circuit.png
```

```python
model = Icm(circuit)
icircuit = model.to_icm()
icircuit.png
```

```python
g = construct_icm_graph(icircuit)
draw_graph(g)
```

```python
g = combine_graph(g, circuit)
draw_graph(g)
```

```python
g = teleport(g)
draw_graph(g)
```

```python
three_loop_reduction(g, 1)
draw_graph(g)
```

```python
g = teleport(g)
draw_graph(g)
```
