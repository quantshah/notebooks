```python
from qutip import QubitCircuit
import networkx as nx
from icm_graph import construct_icm_graph, draw_graph, combine_graph
```

```python
circuit = QubitCircuit(2, reverse_states=False)

circuit.add_gate("CNOT", targets =[1], controls=[0])
circuit.add_gate("CNOT", targets =[0], controls=[1])
circuit.add_gate("CNOT", targets =[1], controls=[0])
circuit.png
```

```python
g = construct_icm_graph(circuit)
draw_graph(g)
```

```python
g2 = combine_graph(g, circuit)
draw_graph(g2)
```

```python
circuit2 = QubitCircuit(3, reverse_states=False)
circuit2.add_gate("CNOT", controls=[0], targets=[2])
circuit2.add_gate("CNOT", controls=[1], targets=[0])
circuit2.add_gate("CNOT", controls=[1], targets=[2])

circuit2.png
```

```python
g2 = construct_icm_graph(circuit2)
draw_graph(g2)
```

```python
g2 = combine_graph(g2, circuit2)
draw_graph(g2)
```
