```python
from qutip import QubitCircuit
import networkx as nx
from qutip.qip.icm import Icm
from icm_graph import (construct_icm_graph, combine_graph,
                       draw_graph, teleport)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
circuit = QubitCircuit(2, reverse_states=False)

circuit.add_gate("CNOT", targets =[1], controls=[0])
circuit.add_gate("CNOT", targets =[0], controls=[1])
circuit.add_gate("CNOT", targets =[1], controls=[0])

model = Icm(circuit)
icircuit = model.to_icm()

g = construct_icm_graph(icircuit)
g = combine_graph(g, circuit)
draw_graph(g)
```

```python
def two_loop_reduction(g, opt1=0, opt2=0, r_loop=None, o_r_loop=None):
    removeables = []
    optionals = {}
    for node in g.node:
        p = g.node[node]["pin"]
#         print(node, p)
        if p is None:
            neighbors = g.neighbors(node)
            print(node, neighbors)
            if len(neighbors) == 2:
                removeables += [node]
                temp_optional = []
                for temp_node in neighbors:
                    if g.node[temp_node]["pin"] is None:
                        temp_optional += [temp_node]
                if len(temp_optional) > 0:
                    optionals[node] = temp_optional
    print(removeables, optionals)
    if len(removeables) > 0:
        
        if r_loop:
            removeable_loop = r_loop
        else:
            removeable_loop = removeables[opt1]
        
        if o_r_loop:
            optional_removeable_loop = o_r_loop
        else:
            optional_removeable_loop = optionals[removeable_loop][opt2]

        neighborsOfRemoveableLoop = g.neighbors(removeable_loop)
        neighborsOfOptionalLoop = g.neighbors(optional_removeable_loop)

#         print("removeable", removeable_loop)
#         print("optionals", optional_removeable_loop)

        
        opt_to_opt_neighbors = []
        for node in neighborsOfOptionalLoop:
            if node != removeable_loop:
                for temp_node in optionals:
                    opt_to_opt_neighbors += [(node, temp_node)]

        opt_to_rm_neighbors = []

        for node in neighborsOfRemoveableLoop:
            if node != optional_removeable_loop:
                for temp_node in optionals:
                    opt_to_rm_neighbors += [(node, temp_node)]

                
        for n1, n2 in opt_to_opt_neighbors:
            if n2 not in g[n1]:
                g.add_edge(n1, n2)
            elif n2 in g[n1]:
                g.remove_edge(n1, n2)
        
        for n1, n2 in opt_to_rm_neighbors:
            if n2 not in g[n1]:
                g.add_edge(n1, n2)
            elif n2 in g[n1]:
                g.remove_edge(n1, n2)
                
        g.remove_nodes_from([removeable_loop, optional_removeable_loop], self_loop=False)

    return g
```

```python
g = two_loop_reduction(g)
draw_graph(g)
```
